```python
import xgboost as xgb
import pickle
import numpy as np
siz = 50000
zzz = 0


def map5eval(preds, dtrain):
    predicted = []
    for i in range(preds.shape[0]):
        predicted.append(preds[i].argsort()[-5:][::-1])
    predicted = np.array(predicted)
    actual = dtrain.get_label()
    metric = 0.
    for i in range(5):
        metric += np.sum(actual==predicted[:,i])/(i+1)
        print("Recommendation",i+1,":",np.sum(actual==predicted[:,i]))
    metric /= actual.shape[0]
    return 'MAP@5', -metric

def map5eval2(preds,actual):
    predicted = []
    for i in range(preds.shape[0]):
        predicted.append(preds[i].argsort()[-5:][::-1])
    predicted = np.array(predicted)
    metric = 0.
    for i in range(5):
        metric += np.sum(actual==predicted[:,i])/(i+1)
        print("Recommendation",i+1,":",np.sum(actual==predicted[:,i]))
    metric /= actual.shape[0]
    return 'MAP@5', metric, False

xclf = xgb.XGBClassifier(
                objective = 'multi:softmax',
                max_depth = 5,
                n_estimators=1000,
                learning_rate=0.1,
                nthread=4,
                subsample=0.7,
                colsample_bytree=0.7,
                min_child_weight = 3,
                silent=False)
# xgbtrain_vect = []
# xgbtrain_op = []
# xgbvalid_vect = []
# xgbvalid_op = []
# X_test = []
# y_test = []

xgbtrain_vect = []
xgbtrain_op = []
xgbvalid_vect = []
xgbvalid_op = []
X_test = []
y_test = []
rftrain_vect = []
rftrain_op = []
import numpy as np
import pickle
with open("xgbtrain_vect","rb") as file:
    xgbtrain_vect = np.array(pickle.load(file))
with open("xgbtrain_op","rb") as file:
    xgbtrain_op = np.array(pickle.load(file))
with open("xgbvalid_vect","rb") as file:
    xgbvalid_vect = np.array(pickle.load(file))
with open("xgbvalid_op","rb") as file:
    xgbvalid_op = np.array(pickle.load(file))
with open("X_test","rb") as file:
    X_test = np.array(pickle.load(file))
with open("y_test","rb") as file:
    y_test = np.array(pickle.load(file))
with open("rftrain_vect","rb") as file:
    rftrain_vect = np.array(pickle.load(file))
with open("rftrain_op","rb") as file:
    rftrain_op = np.array(pickle.load(file))

print("fitting model")
xclf.fit(xgbtrain_vect, xgbtrain_op, eval_metric = map5eval, eval_set=[(xgbvalid_vect,xgbvalid_op)], early_stopping_rounds=50)
exit(0)
print("saving model..")

xclf.save_model("XGB_FINAL_MODEL_FINALLLLLLL3")
# XGB_FINAL_MODEL_FINALLLLLLL - test set as val
# XGB_FINAL_MODEL_FINALLLLLLL2 - xgbval used
# XGB_FINAL_MODEL_FINALLLLLLL3 - rftrain only

print("saved")
print("predicting probabilities..")
xpreds = xclf.predict_proba(X_test)
print("map5:",map5eval2(xpreds,y_test))
xgb_preds = []
with open("XGB_PREDICTIONS_2","wb") as file:
    # xgb_preds = pickle.load(file)
    pickle.dump(xpreds, file)
exit(0)
rf_preds = []
with open("RF_PREDICTIONS_FINALLL","rb") as file:
    rf_preds = pickle.load(file)


'''
[8497]  validation_0-merror:0.6085  validation_0-MAP@5:-0.466092
[8498]  validation_0-merror:0.608267    validation_0-MAP@5:-0.46618
[8499]  validation_0-merror:0.6083  validation_0-MAP@5:-0.466246
'''
```
